```python
# prepare data training 
import pandas as pd 
import numpy as np 


df = {'id':[],'bin_size': [], 'data_cut': [], 'data': []}
num_rows = 400 
for row in range(num_rows):
    bin_size = [np.random.randint(70,100),np.random.randint(70,100),np.random.randint(70,100)]
    data = Generator(n_items= np.random.randint(50,80), seed = row,bin_size= bin_size, n_sample = 0)
    data.generate()
    df['id'].append(row) 
    df['bin_size'].append(bin_size)
    df['data_cut'].append(data.items[0])
    list_item = []
    for item in data.items[0]:
        list_item.append(item[1])
    df['data'].append(list_item)
        

    
```

```python
df_1 = pd.DataFrame(df)
```
